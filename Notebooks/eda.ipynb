# Employee Performance Predictor - EDA

## Objective
Analyze employee dataset to understand patterns affecting performance.

## Features
- Age
- Experience
- Salary
- Department
- Training Hours
- Projects
- Performance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../data/employee_data.csv')
df.head()
```

```python
df.info()
df.describe()
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure()
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='department', data=df)
plt.title('Department Distribution')
plt.show()
```

```python
sns.countplot(x='performance', data=df)
plt.title('Performance Distribution')
plt.show()
```

```python
sns.boxplot(x='performance', y='experience', data=df)
plt.title('Experience vs Performance')
plt.show()
```

```python
sns.boxplot(x='performance', y='training_hours', data=df)
plt.title('Training vs Performance')
plt.show()
```

```python
sns.boxplot(x='performance', y='projects', data=df)
plt.title('Projects vs Performance')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
plt.figure()
sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

## Key Insights

1. Higher experience leads to better performance.
2. Training hours strongly impact performance.
3. Projects influence medium performers.
4. Salary has moderate impact.
5. Department has less impact.
